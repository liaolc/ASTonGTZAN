```python
#DEPRECATED -- Included inside overall GTZAN tar ball


##Adapted from Yuan Gong (2021) -- yuangong@mit.edu

import numpy as np
import json
import os
import wget
# from google.colab import drive
# drive.mount('/content/drive')
# PATH_OF_DATA= '/content/gdrive/MyDrive/GTZAN/Data/'


# prepare the data of the GTZAN dataset.
print('Now download and process GTZAN dataset, it will take a few moments...')
# if os.path.exists('./data') == False:
#     os.mkdir('./data')
# # download the speechcommands dataset
# if os.path.exists('./data/GTZAN') == False:
#     # we use the 35 class v2 dataset, which is used in torchaudio https://pytorch.org/audio/stable/_modules/torchaudio/datasets/speechcommands.html
#     sc_url = 'https://www.dropbox.com/scl/fi/80mng4iut8k16urm58awr/GTZANshort.tar.gz?rlkey=w2hiud7zo92e3foli490rect4&st=mddvegxr&dl=1'
#     wget.download(sc_url, out='./data/')
#     print('Finished Download, Extracting')
#     os.mkdir('./data/GTZAN')
#     os.system('tar -xzvf ./data/GTZANshort.tar.gz -C ./data/GTZAN')
#     #os.remove('./data/GTZAN.tar.gz')
# # Use training list from coreyker - "https://github.com/coreyker/dnn-mgr/tree/master/gtzan"
with open('./egs/GTZAN/data/train_filtered.txt', 'r+') as f:
  train_list = f.readlines()
  extra_list = []
  i = 0
  for count in range(len(train_list)):
    # if count > 1:

    line = train_list[i]
    newline = line.strip()
    parts = line.split('/')
    parts[0] += '2'
    newline = ('/'.join(parts))
    extra_list.append(newline)

    parts[0] = parts[0].replace("2", "3", 1)
    newline = ('/'.join(parts))
    extra_list.append(newline)
    train_list[i+1:i+1] = extra_list
    if i < 10:
      print(i, train_list)
    i+= len(extra_list)+1

    extra_list.clear()
  #train_list += extra_list
  #with open('.data/GTZAN/Data/train_filtered.txt', 'w') as f:
  f.seek(0)
  f.writelines(train_list)
print("Train_list finished processing")
with open('./egs/GTZAN/data/valid_filtered.txt', 'r+') as f:
  val_list = f.readlines()
  extra_list = []
  i = 0
  for count in range(len(val_list)):


    line = val_list[i]

    newline = line.strip()
    parts = line.split('/')
    parts[0] += '2'
    newline = ('/'.join(parts))
    extra_list.append(newline)

    parts[0] = parts[0].replace("2", "3", 1)
    newline = ('/'.join(parts))
    #print(newline)
    extra_list.append(newline)
    val_list[i+1:i+1] = extra_list
    i+= len(extra_list)+1
    extra_list.clear()
  #val_list += extra_list
  #with open('.data/GTZAN/Data/val_filtered.txt', 'w') as f:
  f.seek(0)
  f.writelines(val_list)
print("val_list finished processing")
test_list = []
with open('./egs/GTZAN/data/test_filtered.txt', 'r+') as f:
  test_list = f.readlines()
  i = 0
  extra_list = []
  for count in range(len(test_list)):

    # if count > 1:

    line = test_list[i]
    newline = line.strip()
    parts = line.split('/')
    parts[0] += '2'
    newline = ('/'.join(parts))
    extra_list.append(newline)

    parts[0] = parts[0].replace("2", "3", 1)
    newline = ('/'.join(parts))
    extra_list.append(newline)

    test_list[i+1:i+1] = extra_list
    i+= len(extra_list)+1
    extra_list.clear()
  #test_list += extra_list
  #with open('.data/GTZAN/Data/test_filtered.txt', 'w') as f:
  f.seek(0)
  f.writelines(test_list)
# with open('./data/GTZAN/Data/valid_filtered.txt', 'r+') as f:
#   f.append(test_list)
print("Test_list finished processing")

label_set = np.loadtxt('./egs/GTZAN/data/GTZAN_class_labels_indices.csv', delimiter=',', dtype='str')
label_map = {}
for i in range(1, len(label_set)):
    label_map[eval(label_set[i][2])] = label_set[i][0]
print(label_map)

# generate  json files
if os.path.exists('./egs/GTZAN/data/newdatafiles') == False:
    os.mkdir('./egs/GTZAN/data/newdatafiles')
    base_path = './egs/GTZAN/data/'
    for split in ['test_filtered', 'valid_filtered', 'train_filtered']:
        wav_list = []
        with open(base_path+split+'.txt', 'r') as f:
            filelist = f.readlines()
        i = 0
        for count in filelist:
            file = filelist[i]
            file = file.strip()
            if i < 10:
                print(i, file)
            label_title = file.split('/')[0]

            if "2" in label_title:
                label_title = label_title.replace("2", "")
            if "3" in label_title:
                label_title = label_title.replace("3", "")
            #print(file.split(file.split('/')[0]))
            #print(cur_label)
            #print(label_title)
            cur_label = label_map[label_title]

            #cur_path = os.path.abspath('os.getcwd()) + '/data/GTZAN/Data/'' + file.strip()
            cur_path = '/content/drive/MyDrive/ASTonGTZAN-main/GTZANdatafiles/' + file.strip()
            cur_dict = {"wav": cur_path, "labels": '/m/gtzan'+cur_label.zfill(2)}
            #wav_list[i+1:i+1] = cur_dict
            wav_list.append(cur_dict)
            if i < 10:
              print(i, wav_list)
            i+= 1
        if split == 'train_filtered':
            with open('./egs/GTZAN/data/newdatafiles/GTZAN_train_data.json', 'w') as f:
                json.dump({'data': wav_list}, f, indent=1)
        if split == 'test_filtered':
            with open('./egs/GTZAN/data/newdatafiles/GTZAN_eval_data.json', 'w') as f:
                json.dump({'data': wav_list}, f, indent=1)
        if split == 'valid_filtered':
            with open('./egs/GTZAN/data/newdatafiles/GTZAN_valid_data.json', 'w') as f:
                json.dump({'data': wav_list}, f, indent=1)
        print(split + ' data processing finished, total {:d} samples'.format(len(wav_list)))

    print('GTZAN dataset processing finished.')


```

```python
##Adapted from Yuan Gong

import torch
import numpy as np

import dataloader

# set skip_norm as True only when you are computing the normalization stats
audio_conf = {'num_mel_bins': 128, 'target_length': 1024, 'freqm': 24, 'timem': 192, 'mixup': 0.5, 'skip_norm': True, 'mode': 'train', 'dataset': 'audioset'}

train_loader = torch.utils.data.DataLoader(
    dataloader.AudiosetDataset('/data/sls/scratch/yuangong/audioset/datafiles/balanced_train_data.json', label_csv='/data/sls/scratch/yuangong/audioset/utilities/class_labels_indices.csv',
                                audio_conf=audio_conf), batch_size=1000, shuffle=False, num_workers=8, pin_memory=True)
mean=[]
std=[]
for i, (audio_input, labels) in enumerate(train_loader):
    cur_mean = torch.mean(audio_input)
    cur_std = torch.std(audio_input)
    mean.append(cur_mean)
    std.append(cur_std)
    print(cur_mean, cur_std)
print(np.mean(mean), np.mean(std))
```
